```python
grouped = df.groupby('Season')

def wavg(group):
    d = group['Ratings']
    w = group['Votes']
    return (d * w).sum() / w.sum()

wavg_ratings = grouped.apply(wavg)
```

```python
avg_viewership = df.groupby('Season')['Viewership'].mean()
avg_viewership
```

```python
df.groupby('Season')['Ratings'].mean()
```

```python
wavg_ratings
```

```python
df_rating= pd.DataFrame(wavg_ratings)
```

```python
df_rating
```

```python
df2 = df.merge(df_rating, how = 'inner', left_on='Season', right_on= 'Season')
```

```python
df2 = df2.rename(columns = {0 : 'Avg Rating'})
```

```python
df2
```

```python
df2.groupby('Season')['Season'].mean()
```

```python
seasons = range(1,10)
rating = df2.groupby('Season')['Avg Rating'].mean()
```

```python
plt.bar(seasons, rating)
xticks = seasons
plt.ylim(5, 10)
plt.title("Rating for each season")
plt.xlabel('Season')
plt.ylabel('Rating')
plt.show()
```
